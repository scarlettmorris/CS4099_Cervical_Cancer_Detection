```python
import os
import zarr
```

```python
zarr_file_path = "data/embeddings/IC-CX-00001-01.zarr"

if not os.path.exists(zarr_file_path):
    print(f"File does not exist: {zarr_file_path}")
else:
    print(f"File exists: {zarr_file_path}")
```

```python
# Path to a single .zarr file
zarr_file_path = "data/embeddings/IC-CX-00001-01.zarr"
zarr_data = zarr.open(zarr_file_path, mode='r')

# Inspect the structure of the zarr file
print(zarr_data.tree())
features = zarr_data['features'][:]
filenames = zarr_data['filenames'][:]
```
